### It shows that the dataset consist of crime records for the past 22 years. For this particular analyses, i'll be focusing on the past 10 years in order to create an accurate and relevant representation of the phenomenom. By this, I will also be able to identify any recent pattern or trend in the dataset. 

```python
# crop dataset to the past 10 years

df = data[data['Year'] > 2012]

df1 = df.copy()
```

### Data Cleaning

```python
# data cleaning 
df1.isnull().sum()
```

```python
# drop redundant variable 

df1.drop(columns= ['Location','Ward','Beat','District','Community Area','FBI Code','Unnamed: 0','Case Number','X Coordinate','Y Coordinate','Updated On','Latitude','Longitude','IUCR'], inplace= True)
```

```python
# Investigating the cropped dataframe after dropping redundant variables

df1.info()
df1.head()
```

```python
# fill null values in the location description 

df1['Location Description'].fillna('OTHERS', inplace= True)
```

```python
# check dataframe information

df1.info()
```

```python
# created new column called time of the from the date timestamp

df1['time_of_day'] = df1['Date'].apply(lambda x: 'Morning' if 'AM' in str(x) else 'Evening')
```

```python
# check cleaned dataframe information 

df1.info()
```

```python
# quick overview of the cleaned dataframe

df1.head()
```

```python
# write cleaned dataframe to a csv file

df1.to_csv("new_chicago_crime_data.csv", index=False)
```
